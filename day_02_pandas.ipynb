```python
import pandas as pd 
import numpy as np 

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
```

```python
df.isnull()
```

```python
missing_count = df.isnull().sum()
missing_percent = df.isnull().sum() / df.shape[0] * 100

missing_table = pd.DataFrame({
    'missing': missing_count,
    'missing_%': missing_percent.round(2)
})

missing_table.sort_values('missing', ascending=False)

```

```python
df['is_alone'] = df['family_size'] == 1
```

```python
df.groupby('is_alone')['Survived'].mean()
```

```python
df['Age_filled'] = df['Age'].fillna(df['Age'].median()).round(0)
```
