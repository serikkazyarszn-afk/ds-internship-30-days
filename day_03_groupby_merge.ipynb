```python
import pandas as pd 
import seaborn as sns 

df = sns.load_dataset('tips')
```

```python
df['tip_percent'] = df['tip'] / df['total_bill'] * 100
df
```

```python
df.groupby('day')['tip_percent'].mean().round(2).reset_index()
```

```python
# 1. Какой средний tip_percent по дням недели?
#    Average tip percentage per day

# 2. Какой ДЕНЬ + ВРЕМЯ даёт самые высокие чаевые в процентах?
#    Highest average tip_percent for (day × time) combination.
#    Используй groupby с двумя колонками

# 3. Посчитай одной командой по дням: count, mean, median, std чаевых
#    Multiple aggregations at once

# 4. Для каждого дня: общий оборот (sum total_bill), средний чек и количество визитов.
#    Используй named aggregation (самый красивый синтаксис):
#    df.groupby('day').agg(
#        revenue=('total_bill', 'sum'),
#        avg_bill=('total_bill', 'mean'),
#        visits=('size', 'count')
#    )

# 5. Создай новую колонку `day_avg_tip` — средний tip_percent в день визита.
#    Для каждой строки — своё значение.
#    Подсказка: groupby + transform
```

```python
(df.groupby(['day', 'time'])['tip_percent']
    .mean()
    .round(2)
    .sort_values(ascending=False)
    .reset_index())
```

```python
(df.groupby(['day', 'time'])['tip_percent']
    .mean()
    .round(2)
    .idxmax())
```

```python
# 6. Создай искусственную вторую таблицу:
day_info = pd.DataFrame({
    'day': ['Thur', 'Fri', 'Sat', 'Sun'],
    'is_weekend': [False, False, True, True],
    'day_rank': [4, 5, 6, 7]
})
day_info
# Объедини df с day_info через merge по колонке 'day'. Используй LEFT join.
# Merge df with day_info using LEFT join on 'day'.

pd.merge(df, day_info, on='day', how='left')
# Проверь: df.shape до и после — должно остаться столько же строк


# 7. Посчитай средний tip_percent для выходных vs будней
# Используй новую колонку is_weekend из задачи 6
```

```python
df = pd.merge(df, day_info, on='day', how='left')

df.groupby('is_weekend').agg(
    avg_tip_percent=('tip_percent', 'mean'),
    visits=('tip_percent', 'count')
).round(2)
```

```python
df['segment'] = pd.cut(
    df['tip_percent'],
    bins=[0, 15, 20, 100],
    labels=['low', 'medium', 'high']
)
```

```python
high = df[df['segment'] == 'high']
```

```python
high['sex'].value_counts(normalize=True) * 100
high['day'].value_counts(normalize=True) * 100
high['time'].value_counts(normalize=True) * 100
high['smoker'].value_counts(normalize=True) * 100
```
